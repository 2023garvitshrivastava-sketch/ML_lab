```python
# ================================================
# ðŸ§  Hebbian Learning Demonstration (Numerical)
# ================================================

import numpy as np
import matplotlib.pyplot as plt

# --- 1. Define dataset ---
# Simple input patterns (2D)
X = np.array([
    [1, 1],
    [1, -1],
    [-1, 1],
    [-1, -1]
])

# Target output (we'll make it learn to output positive for similar signs)
y_target = np.array([1, -1, -1, 1])

# --- 2. Initialize weights ---
np.random.seed(42)
w = np.random.randn(2) * 0.1
eta = 0.1  # learning rate
epochs = 10

# To record weight evolution
w_history = [w.copy()]

# --- 3. Hebbian learning loop ---
for epoch in range(epochs):
    for i in range(len(X)):
        x = X[i]
        y = np.dot(w, x)              # neuron output
        w += eta * x * y              # Hebbian update rule
    w_history.append(w.copy())
    print(f"Epoch {epoch+1}: weights = {w}")

# --- 4. Convert weight history to array ---
w_history = np.array(w_history)

# --- 5. Visualization ---
plt.figure(figsize=(8,5))
plt.plot(w_history[:,0], label='w1')
plt.plot(w_history[:,1], label='w2')
plt.xlabel("Epoch")
plt.ylabel("Weight Value")
plt.title("Hebbian Learning Weight Update over Epochs")
plt.legend()
plt.grid(True)
plt.show()

# --- 6. Test the learned weights ---
print("\nTesting the final weights:")
for i, x in enumerate(X):
    output = np.dot(w, x)
    print(f"Input: {x}, Output: {output:.3f}, Target: {y_target[i]}")
```
